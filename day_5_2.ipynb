```python
import numpy as np
def modes(code):
    inm = np.full(4, 0)
    if len(code) > 2:
        for x in range(len(code) - 2, 0, -1):
            if code[x - 1] == "1":
                inm[len(code) - x - 2 + 1] = 1

    return inm
```

```python
def parameters1(v, p, m):
    p1 = v[p+1]
    if m[1] == 0:
        p1 = v[p1]

    return p1

def parameters2(v, p, m):
    p1 = v[p+1]
    if m[1] == 0:
        p1 = v[p1]

    p2 = v[p+2]
    if m[2] == 0:
        p2 = v[p2]
        
    return [p1, p2]
```

```python
p = 0
out = 0

while(v[p] != 99):
    
    code = str(v[p])
    instruction = int(code[len(code) - 2:])
    m = modes(code)
        
    if instruction == 1: # add
        p1, p2 = parameters2(v, p, m)
                
        ### Parameters that an instruction writes to will never be in immediate mode.    
        v[v[p+3]] = p1 + p2
        
        p += 4
        
    elif instruction == 2: # mul
        p1, p2 = parameters2(v, p, m)

        ### Parameters that an instruction writes to will never be in immediate mode.    
        v[v[p+3]] = p1 * p2
        
        p += 4

    elif instruction == 3: # input
        iv = inputValues[inputCounter]
        inputCounter += 1
        v[v[p+1]] = iv
        
        p += 2
        
    elif instruction == 4: # output  
        out = parameters1(v, p, m)
        print(out)
        
        p += 2
        
    elif instruction == 5: # jump if true
        p1, p2 = parameters2(v, p, m)
            
        if p1 != 0:
            p = p2
        else:
            p += 3
        
    elif instruction == 6: # jump if false
        p1, p2 = parameters2(v, p, m)
            
        if p1 == 0:
            p = p2
        else:
            p += 3

    elif instruction == 7: # less than
        p1, p2 = parameters2(v, p, m)
            
        if p1 < p2:
            v[v[p+3]] = 1
        else:
            v[v[p+3]] = 0
            
        p += 4

    elif instruction == 8: # equal
        p1, p2 = parameters2(v, p, m)
            
        if p1 == p2:
            v[v[p+3]] = 1
        else:
            v[v[p+3]] = 0
            
        p += 4
        
    else:
        print("error", p)

#assert out == 13547311
```

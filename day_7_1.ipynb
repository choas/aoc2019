```python
import numpy as np
def modes(code):
    inm = np.full(4, 0)
    if len(code) > 2:
        for x in range(len(code) - 2, 0, -1):
            if code[x - 1] == "1":
                inm[len(code) - x - 2 + 1] = 1

    return inm
```

```python
def parameters1(v, p, m):
    p1 = v[p+1]
    if m[1] == 0:
        p1 = v[p1]

    return p1

def parameters2(v, p, m):
    p1 = v[p+1]
    if m[1] == 0:
        p1 = v[p1]

    p2 = v[p+2]
    if m[2] == 0:
        p2 = v[p2]
        
    return [p1, p2]
```

```python
import numpy as np
def modes(code):
    #print(code)
    inm = np.full(4, 0)
    if len(code) > 2:
        for x in range(len(code) - 2, 0, -1):
            if code[x - 1] == "1":
                inm[len(code) - x - 2 + 1] = 1

            if code[x - 1] == "2":
                inm[len(code) - x - 2 + 1] = 2


    return inm
```

```python
def parameters1(v, p, m, rel):
    #print(m)    
    p1 = v[p+1]
    if m[1] == 0:
        p1 = v[p1]
    elif m[1] == 2:
        print("**", p1, m, rel)
        p1 = v[p1 + rel]

    return p1

def parameters2(v, p, m, rel):
    #print(m)
    p1 = v[p+1]
    if m[1] == 0:
        p1 = v[p1]
    elif m[1] == 2:
        print("**2.1", p1, m, rel)
        p1 = v[p1 + rel]

    p2 = v[p+2]
    if m[2] == 0:
        p2 = v[p2]
    elif m[2] == 2:
        print("**2.2", p1, m, rel)
        p2 = v[p2 + rel]

    return [p1, p2]
```

```python
class Amp:
    p = 0
    #out = inputValues[len(inputValues) - 1]
    inputCounter = 0


    def __init__(self, v, inputValues):
        self.v = v
        self.inputValues = inputValues
        self.inputCounter = 0
        self.p = 0
        self.halt = False
        self.rel = 0
    
    def run(self, inputValue):
        
        #print(self.v)
        
        #print(">", inputValue)
        if self.halt:
            print("H", self.inputValues[0])
        
        self.inputValues.append(inputValue)
        
        while(self.v[self.p] != 99):

            code = str(self.v[self.p])
            instruction = int(code[len(code) - 2:])
            m = modes(code)

            if instruction == 1: # add
                p1, p2 = parameters2(self.v, self.p, m, self.rel)

                if m[3] == 2:
                    self.v[self.v[self.p+3] + self.rel] = p1 + p2
                else:
                    self.v[self.v[self.p+3]] = p1 + p2

                self.p += 4

            elif instruction == 2: # mul
                print("CODE:", code, m)
                p1, p2 = parameters2(self.v, self.p, m, self.rel)

                if m[3] == 2:
                    self.v[self.v[self.p+3] + self.rel] = p1 * p2
                else:
                    self.v[self.v[self.p+3]] = p1 * p2

                self.p += 4

            elif instruction == 3: # input
                #if self.inputCounter < len(self.inputValues):
                iv = self.inputValues[self.inputCounter]

                #else:
                #    iv = self.inputValues[len(self.inputValues) - 1]

                self.inputCounter += 1
                #print("<", iv, self.inputCounter, self.p)
                
                if m[1] == 2:
                    print("IN m", m, iv)
                    self.v[self.v[self.p+1] + self.rel] = iv
                else:
                    self.v[self.v[self.p+1]] = iv

                self.p += 2

            elif instruction == 4: # output  
                iv = parameters1(self.v, self.p, m, self.rel)
                #print(">", iv, self.inputValues[0], self.p, code)
                #print("V", self.v)
                self.out = iv #.append(iv)
                self.p += 2

                return self.out, False


            elif instruction == 5: # jump if true
                p1, p2 = parameters2(self.v, self.p, m, self.rel)

                if p1 != 0:
                    self.p = p2
                else:
                    self.p += 3

            elif instruction == 6: # jump if false
                p1, p2 = parameters2(self.v, self.p, m, self.rel)

                if p1 == 0:
                    self.p = p2
                else:
                    self.p += 3

            elif instruction == 7: # less than
                p1, p2 = parameters2(self.v, self.p, m, self.rel)

                r = 0
                if m[3] == 2:
                    r = self.rel
                
                if p1 < p2:
                    self.v[self.v[self.p+3] + r] = 1
                else:
                    self.v[self.v[self.p+3] + r] = 0

                self.p += 4

            elif instruction == 8: # equal
                p1, p2 = parameters2(self.v, self.p, m, self.rel)

                r = 0
                if m[3] == 2:
                    r = self.rel

                if p1 == p2:
                    self.v[self.v[self.p+3] + r] = 1
                else:
                    self.v[self.v[self.p+3] + r] = 0

                self.p += 4

            elif instruction == 9: # adjust relative base
                
                p1 = parameters1(self.v, self.p, m, self.rel)
                print("REL", p1)
                self.rel += p1
                
                self.p += 2


            else:
                print("error", self.p)
        self.halt = True
        #print("*", self.out, self.inputValues[0])
        return [self.out, True]

    def getV(self):
        return self.v
```

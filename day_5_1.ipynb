```python
import numpy as np
def modes(code):
    inm = np.full(4, 0)
    if len(code) > 2:
        for x in range(len(code) - 2, 0, -1):
            if code[x - 1] == "1":
                inm[len(code) - x - 2 + 1] = 1

    return inm
```
